 ## Mini Project 1: Fizz Buzz


```python

for number in range(1, 101):

    if number % 3 == 0 and number % 5 == 0:
        
        print("Fizz Buzz")
    elif number % 3 == 0:
        print("Fizz")
    elif number % 5 == 0:
        print("Buzz")
    else:
        print(number)
```

## Mini Project 2 â€“ Movie Budget Analysis




```python
movies = [
    ("Eternal Sunshine of the Spotless Mind", 20000000),
    ("Memento", 9000000),
    ("Requiem for a Dream", 4500000),
    ("Pirates of the Caribbean: On Stranger Tides", 379000000),
    ("Avengers: Age of Ultron", 365000000),
    ("Avengers: Endgame", 356000000),
    ("Incredibles 2", 200000000)
]

total = 0
for m in movies:
    total += m[1]

average = total / len(movies)
print("Average Budget:", average)

print("\nMovies with budget higher than average:\n")

count = 0
for name, budget in movies:
    if budget > average:
        print(name, "-", (budget - average), "higher than average")
        count += 1

print("\nTotal Movies:", len(movies))
print("Movies above average:", count)
```
